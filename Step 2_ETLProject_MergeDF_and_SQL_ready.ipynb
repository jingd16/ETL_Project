```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
mens_csv_path = "Mens_Shoes_df.csv"
womens_csv_path = "Womens_Shoes_df.csv"

#Import the CSV into a pandas DataFrame
mens_df = pd.read_csv(mens_csv_path)
Appended_df = mens_df.append(pd.read_csv(womens_csv_path))
```

```python
# Clean up Brands column by converting all values to Title Case 

NumBrandsBefore = Appended_df['Brand'].nunique()
Appended_df['Brand'] = Appended_df['Brand'].str.title()
NumBrandsAfter = Appended_df['Brand'].nunique()
print(f'Num Brands before clean: {NumBrandsBefore} and Num Brands after clean: {NumBrandsAfter}')
```

```python
# Clean up Colours column by converting all values to Title Case 

NumColoursBefore = Appended_df['Colours'].nunique()
Appended_df['Colours'] = Appended_df['Colours'].str.title()
NumColoursAfter = Appended_df['Colours'].nunique()
print(f'Num Colours before clean: {NumColoursBefore} and Num Colours after clean: {NumColoursAfter}')
```

```python
# Drop currency and index column 
Appended_df.drop('Currency', inplace = True, axis = 1)
Appended_df.drop('Index', inplace = True, axis = 1)
```

```python
Appended_df
```

```python
# Create Brand table


#Convert brand from each dataset to a list
brand_name = Appended_df["Brand"].tolist()


#Drop off duplicated brand names
from collections import OrderedDict
brand_name = list(OrderedDict.fromkeys(brand_name))
brand_name_sorted = brand_name.sort()

#Create an Brand_ID list
ID=range(1,len(brand_name)+1)
brand_name_ID=[]
for x in ID:
    brand_name_ID.append("BN"+str(x))

#Create an Brand_Name dataframe
Brand_Name_df = pd.DataFrame({"Brand_ID": brand_name_ID, 
                              "Brand_Name" :brand_name}, index=None)

Brand_Name_df.set_index('Brand_ID')
```

```python
#===============================CREATE COLOR TABLE ========================




#Convert brand from each dataset to a list

color = Appended_df["Colours"].tolist()

#rename list 
combined_color = color

#Drop off duplicated brand names
from collections import OrderedDict
combined_color = list(OrderedDict.fromkeys(combined_color))

#Create an Brand_ID list
ID=range(1,len(combined_color)+1)
Color_ID=[]
for x in ID:
    Color_ID.append("CL"+str(x))

#Create an Brand_Name dataframe
Color_df = pd.DataFrame({"Color_ID": Color_ID, 
                              "Colors" :combined_color}, index=None)

Color_df
```

```python
# Copy "Appended_df" to Shoes_df

Shoes_df = Appended_df.copy()
Shoes_df
```

```python
# Merge the Brand and Shoe tables to place Brand_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, Brand_Name_df, left_on="Brand", right_on="Brand_Name", how='left')

# Drop redundant Brand Name columns 
Shoes_df.drop('Brand', inplace = True, axis = 1)
Shoes_df.drop('Brand_Name', inplace = True, axis = 1)

Shoes_df
```

```python
# Merge the Colours and Shoe tables to place Color_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, Color_df, left_on="Colours", right_on='Colors', how='left')

# # Drop redundant Colour Name columns 
Shoes_df.drop('Colours', inplace = True, axis = 1)
Shoes_df.drop('Colors', inplace = True, axis = 1)
```

```python
Shoes_df
```

```python
# Manually Create Gender table - tiny amount of data 

gender_dictionary = {'Gender': ["Men's","Women's"],
        'Gender_ID': ["M","W"]
        }

gender_df = pd.DataFrame(gender_dictionary, columns = ['Gender', 'Gender_ID'])

gender_df
```

```python
# Merge the Gender and Shoe tables to place Gender_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, gender_df, left_on="Gender", right_on='Gender')

# Drop the redundant Gender column
Shoes_df.drop('Gender', inplace = True, axis = 1)
```
