```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
#Create reference to CSV file
csv_path = "Datafiniti_Mens_Shoe_Prices.csv"

#Import the CSV into a pandas DataFrame
Mens_Shoes_df = pd.read_csv(csv_path, low_memory = False)
Mens_Shoes_df
```

```python
#Show the 'shape' of the dataframe 
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

#### COLLECT ONLY REQUIRED COLUMNS

```python
#Select only columns we need
Mens_Shoes_df = Mens_Shoes_df[["id", "brand", "colors", "name", "prices.amountMax",  "prices.currency", "prices.dateSeen",
                               "prices.isSale"]]

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### REMOVE ALL ROWS WHICH CONTAIN A CURRENCY OTHER THAN USD

```python
#Get names of indexes for which column Currency has value other than USD
Mens_Shoes_df_indexnames = (Mens_Shoes_df[Mens_Shoes_df['prices.currency'] != "USD"]).index

#Delete these row indexes from dataframe
Mens_Shoes_df.drop(Mens_Shoes_df_indexnames, inplace=True)

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### REMOVE ALL ROWS WITH FALSE IN THE SALE COLUMN

```python
#Get names of indexes for which column Currency has value other than USD
Mens_Shoes_df_indexnames = (Mens_Shoes_df[Mens_Shoes_df['prices.isSale'] == "FALSE"]).index

#Delete these row indexes from dataFrame
Mens_Shoes_df.drop(Mens_Shoes_df_indexnames, inplace=True)

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

```python
#Drop the 'prices.isSale' column as it is now useless data - was used to siphon the non-Sale prices.
Mens_Shoes_df = Mens_Shoes_df.drop(['prices.isSale'], axis = 1)

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### FILTER THE DATE COLUMN TO CORRECT FORMAT(YYYY-MM-DD) AND REMOVE ANY 'EMPTY' COLUMN

```python
#Filter out non-DateTime valued rows, and set value to "NaN"
Mens_Shoes_df['prices.dateSeen'] = pd.to_datetime(Mens_Shoes_df['prices.dateSeen'], errors = 'coerce')

#Drop all "NaN" valued rows in prices.dateSeen column
Mens_Shoes_df = Mens_Shoes_df.dropna(subset = ['prices.dateSeen'])

#Convert the price.dateSeen column to string (from TimeStamp due to previous function to_datetime used)
Mens_Shoes_df['prices.dateSeen'] = Mens_Shoes_df['prices.dateSeen'].astype(str)

#Convert column to list
dt_list = Mens_Shoes_df["prices.dateSeen"].tolist()

#Clean up the list to only show yyyy-mm-dd formate
w_date = []
for x in dt_list:
    w_date.append(x.split(" ")[0])

#Add the formated list into the dataframe as new column "Date"
Mens_Shoes_df["Date"] = w_date

#Drop all "NaN" valued rows in prices.dateSeen column
Mens_Shoes_df = Mens_Shoes_df.dropna(subset=['Date'])

#Drop the 'prices.dateSeen' column as it is now repleced with a 'Date' column
Mens_Shoes_df = Mens_Shoes_df.drop(['prices.dateSeen'], axis = 1)

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### FILTER THE NAME COLUMN TO COLLECT REQUIRED ROWS

```python
#Define shoe list to filter out non-shoe items 
ShoeList = ['Sandals', 'Sandal', 'Shoe', 'Shoes', 'Boots', 'Boot', 'Sneakers', 'Sneaker', 'Loafers', 'Slippers', 
            'Runners', 'Trainers', 'Thongs', 'bootie', 'wedge', 'heel']

# Search for content in ShoeList and filter 
Mens_Shoes_df = Mens_Shoes_df[Mens_Shoes_df.name.str.contains('|'.join(ShoeList), case=False)]

#Drop all "NaN" value rows in the name column
Mens_Shoes_df = Mens_Shoes_df.dropna(subset=['name'])

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### FILTER THE COLOR COLUMN TO COLLECT REQUIRED ROWS

```python
#Replace "," with "/", so we can split strings all together
colour_list = Mens_Shoes_df.colors.str.replace(',', '/')

#Loop through the list and pick out the first color before "/"
colour=[]
for x in colour_list:
    colour.append(str(x).split("/")[0])
    
#Create a new column with the simplified color
Mens_Shoes_df["Colours"] = colour

#Drop the 'color' column as it is now repleced with a 'Colour' column
Mens_Shoes_df = Mens_Shoes_df.drop(['colors'], axis = 1)

#Drop off rows with no brand
Mens_Shoes_df = Mens_Shoes_df.dropna(subset=['Colours'])

#Get names of indexes for which the Colours column has text containing "nan"
Mens_Shoes_indexNames = Mens_Shoes_df[Mens_Shoes_df['Colours'] == "nan"].index
Mens_Shoes_df.drop(Mens_Shoes_indexNames, inplace=True)

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### REMOVE ALL ROWS WITH NO DATA

```python
#Filter out those rows which does not contain any data 
Mens_Shoes_df = Mens_Shoes_df.dropna() 

#Confirm rows are dropped
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### FILTER THE DATAFRAME AND DROP DUPLICATES, AND RE-ORDER

```python
# Mens DF - Select Highest Price ID and Drop Duplicates 

# convert prices column to Float 
Mens_Shoes_df['prices.amountMax'] = pd.to_numeric(Mens_Shoes_df['prices.amountMax'], errors = 'coerce')

# sort DF based on Price
Mens_Shoes_df = Mens_Shoes_df.sort_values('prices.amountMax', ascending=False)

# Remove Duplicate ID rows 
Mens_Shoes_df.drop_duplicates(subset=['id'], inplace = True, keep = 'first')

#Confirm rows are dropped
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### ADD A COLUMN TO THE DATAFRAME AND FILL WITH 'Mens'

```python
#Create an Gender column for Men's Shoes
Mens_Shoes_df['Gender'] = "Men's"

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Mens_Shoes_df
```

#### RENAME COLUMNS AND CORRECT THE INDEX FOR VISUAL PURPOSES

```python
#The original dataframe had an index, so re-arrange the index 
Mens_Shoes_df.index = [x for x in range(1, len(Mens_Shoes_df.values)+1)]
Mens_Shoes_df.index.name = 'Index'
```

```python
#Rename the columns for visual purposes
Mens_Shoes_df = Mens_Shoes_df.rename(columns = {"id":"ID", "brand":"Brand", "name":"Name", "prices.amountMax":"Price($)", 
                                                "prices.currency":"Currency", "prices.dateSeen":"Date"})

#Reorganise the columns for visualisation
Mens_Shoes_df = Mens_Shoes_df[['ID', 'Name', 'Brand', 'Colours', 'Price($)', 'Gender', 'Currency', 'Date']]

#Show the 'shape' of the dataframe
shape = Mens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Save dataframe as csv file
Mens_Shoes_df.to_csv('Mens_Shoes_df.csv') 
```

```python
#Create reference to CSV file
mens_csv_path = "Mens_Shoes_df.csv"
womens_csv_path = "Womens_Shoes_df.csv"

#Import the CSV into a pandas DataFrame
mens_df = pd.read_csv(mens_csv_path)
Appended_df = mens_df.append(pd.read_csv(womens_csv_path))
```

```python
# Clean up Brands column by converting all values to Title Case 

NumBrandsBefore = Appended_df['Brand'].nunique()
Appended_df['Brand'] = Appended_df['Brand'].str.title()
NumBrandsAfter = Appended_df['Brand'].nunique()
print(f'Num Brands before clean: {NumBrandsBefore} and Num Brands after clean: {NumBrandsAfter}')
```

```python
# Clean up Colours column by converting all values to Title Case 

NumColoursBefore = Appended_df['Colours'].nunique()
Appended_df['Colours'] = Appended_df['Colours'].str.title()
NumColoursAfter = Appended_df['Colours'].nunique()
print(f'Num Colours before clean: {NumColoursBefore} and Num Colours after clean: {NumColoursAfter}')
```

```python
# Drop currency and index column 
Appended_df.drop('Currency', inplace = True, axis = 1)
Appended_df.drop('Index', inplace = True, axis = 1)
```

```python
Appended_df
```

```python
# Create Brand table


#Convert brand from each dataset to a list
brand_name = Appended_df["Brand"].tolist()


#Drop off duplicated brand names
from collections import OrderedDict
brand_name = list(OrderedDict.fromkeys(brand_name))
brand_name_sorted = brand_name.sort()

#Create an Brand_ID list
ID=range(1,len(brand_name)+1)
brand_name_ID=[]
for x in ID:
    brand_name_ID.append("BN"+str(x))

#Create an Brand_Name dataframe
Brand_Name_df = pd.DataFrame({"Brand_ID": brand_name_ID, 
                              "Brand_Name" :brand_name}, index=None)

Brand_Name_df.set_index('Brand_ID')
```

```python
# CREATE COLOR TABLE
#Convert brand from each dataset to a list

color = Appended_df["Colours"].tolist()

#rename list 
combined_color = color

#Drop off duplicated brand names
from collections import OrderedDict
combined_color = list(OrderedDict.fromkeys(combined_color))

#Create an Brand_ID list
ID=range(1,len(combined_color)+1)
Color_ID=[]
for x in ID:
    Color_ID.append("CL"+str(x))

#Create an Brand_Name dataframe
Color_df = pd.DataFrame({"Color_ID": Color_ID, 
                              "Colors" :combined_color}, index=None)

Color_df
```

```python
# Copy "Appended_df" to Shoes_df

Shoes_df = Appended_df.copy()
Shoes_df
```

```python
# Merge the Brand and Shoe tables to place Brand_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, Brand_Name_df, left_on="Brand", right_on="Brand_Name", how='left')

# Drop redundant Brand Name columns 
Shoes_df.drop('Brand', inplace = True, axis = 1)
Shoes_df.drop('Brand_Name', inplace = True, axis = 1)

Shoes_df
```

```python
# Merge the Colours and Shoe tables to place Color_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, Color_df, left_on="Colours", right_on='Colors', how='left')

# # Drop redundant Colour Name columns 
Shoes_df.drop('Colours', inplace = True, axis = 1)
Shoes_df.drop('Colors', inplace = True, axis = 1)
```

```python
Shoes_df
```

```python
# Manually Create Gender table - tiny amount of data 

gender_dictionary = {'Gender': ["Men's","Women's"],
        'Gender_ID': ["M","W"]
        }

gender_df = pd.DataFrame(gender_dictionary, columns = ['Gender', 'Gender_ID'])

gender_df
```

```python
# Merge the Gender and Shoe tables to place Gender_ID into Shoe table
Shoes_df = pd.merge(Shoes_df, gender_df, left_on="Gender", right_on='Gender')

# Drop the redundant Gender column
Shoes_df.drop('Gender', inplace = True, axis = 1)
```

```python
#===============================SQL Storage==========================
```

```python
#-------Renamed column "Price($)" to "Price" in order to import to SQL
Shoes_df = Shoes_df.rename(columns={'Price($)': 'Price'})

# make ID to be index
Shoes_df.set_index(["ID"])
```

```python
#In order to import to SQL, change all coloumns to lowercase
gender_df.columns = gender_df.columns.str.lower()
Color_df.columns = Color_df.columns.str.lower()
Brand_Name_df.columns = Brand_Name_df.columns.str.lower()
Shoes_df.columns = Shoes_df.columns.str.lower()
```

```python
# Connect to SQL
rds_connection_string = "postgres:" + password + "@localhost:5432/ETC_Project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check for table names
engine.table_names()
```

```python
# Load dataframes into database
# Given we only have 1 set of data for each table, set if_exists to replace
Color_df.to_sql(name='color', con=engine, if_exists='replace', index=True)
Brand_Name_df.to_sql(name="brand", con=engine, if_exists='replace', index=True)
gender_df.to_sql(name="gender", con=engine, if_exists='replace', index=True)
Shoes_df.to_sql(name="shoes", con=engine, if_exists='replace', index=True)
```
