```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
csv_path = "Datafiniti_Womens_Shoes_Jun19.csv"

#Import the CSV into a pandas DataFrame
Womens_Shoes_df = pd.read_csv(csv_path, low_memory = False)
Womens_Shoes_df
```

```python
#Show the 'shape' of the dataframe 
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

#### COLLECT ONLY REQUIRED COLUMNS

```python
#Select only columns we need
Womens_Shoes_df = Womens_Shoes_df[["id", "brand", "colors", "name", "prices.amountMax",  "prices.currency", "prices.dateSeen",
                               "prices.isSale"]]

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### REMOVE ALL ROWS WHICH CONTAIN A CURRENCY OTHER THAN USD

```python
#Get names of indexes for which column Currency has value other than USD
Womens_Shoes_df_indexnames = (Womens_Shoes_df[Womens_Shoes_df['prices.currency'] != "USD"]).index

#Delete these row indexes from dataframe
Womens_Shoes_df.drop(Womens_Shoes_df_indexnames, inplace=True)

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### REMOVE ALL ROWS WITH FALSE IN THE SALE COLUMN

```python
#Get names of indexes for which column Currency has value other than USD
Womens_Shoes_df_indexnames = (Womens_Shoes_df[Womens_Shoes_df['prices.isSale'] == "FALSE"]).index

#Delete these row indexes from dataFrame
Womens_Shoes_df.drop(Womens_Shoes_df_indexnames, inplace=True)

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

```python
#Drop the 'prices.isSale' column as it is now useless data - was used to siphon the non-Sale prices.
Womens_Shoes_df = Womens_Shoes_df.drop(['prices.isSale'], axis = 1)
Womens_Shoes_df

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### FILTER THE DATE COLUMN TO CORRECT FORMAT(YYYY-MM-DD) AND REMOVE ANY 'EMPTY' COLUMN

```python
#Filter out non-DateTime valued rows, and set value to "NaN"
Womens_Shoes_df['prices.dateSeen'] = pd.to_datetime(Womens_Shoes_df['prices.dateSeen'], errors = 'coerce')

#Drop all "NaN" valued rows in prices.dateSeen column
Womens_Shoes_df = Womens_Shoes_df.dropna(subset = ['prices.dateSeen'])

#Convert the price.dateSeen column to string (from TimeStamp due to previous function to_datetime used)
Womens_Shoes_df['prices.dateSeen'] = Womens_Shoes_df['prices.dateSeen'].astype(str)

#Convert column to list
dt_list = Womens_Shoes_df["prices.dateSeen"].tolist()

#Clean up the list to only show yyyy-mm-dd formate
w_date = []
for x in dt_list:
    w_date.append(x.split(" ")[0])

#Add the formated list into the dataframe as new column "Date"
Womens_Shoes_df["Date"] = w_date

#Drop all "NaN" valued rows in prices.dateSeen column
Womens_Shoes_df = Womens_Shoes_df.dropna(subset=['Date'])

#Drop the 'prices.dateSeen' column as it is now repleced with a 'Date' column
Womens_Shoes_df = Womens_Shoes_df.drop(['prices.dateSeen'], axis = 1)

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### FILTER THE NAME COLUMN TO COLLECT REQUIRED ROWS

```python
#Define shoe list to filter out non-shoe items 
ShoeList = ['Sandals', 'Sandal', 'Shoe', 'Shoes', 'Boots', 'Boot', 'Sneakers', 'Sneaker', 'Loafers', 'Slippers', 
            'Runners', 'Trainers', 'Thongs', 'bootie', 'wedge', 'heel']

# Search for content in ShoeList and filter 
Womens_Shoes_df = Womens_Shoes_df[Womens_Shoes_df.name.str.contains('|'.join(ShoeList), case=False)]

#Drop all "NaN" value rows in the name column
Womens_Shoes_df = Womens_Shoes_df.dropna(subset=['name'])

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### FILTER THE COLOR COLUMN TO COLLECT REQUIRED ROWS

```python
#Replace "," with "/", so we can split strings all together
colour_list = Womens_Shoes_df.colors.str.replace(',', '/')

#Loop through the list and pick out the first color before "/"
colour=[]
for x in colour_list:
    colour.append(str(x).split("/")[0])
    
#Create a new column with the simplified color
Womens_Shoes_df["Colours"] = colour

#Drop the 'color' column as it is now repleced with a 'Colour' column
Womens_Shoes_df = Womens_Shoes_df.drop(['colors'], axis = 1)

#Drop off rows with no brand
Womens_Shoes_df = Womens_Shoes_df.dropna(subset=['Colours'])

#Get names of indexes for which the Colours column has text containing "nan"
Womens_Shoes_df_indexNames = Womens_Shoes_df[Womens_Shoes_df['Colours'] == "nan"].index
Womens_Shoes_df.drop(Womens_Shoes_df_indexNames, inplace=True)

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### REMOVE ALL ROWS WITH NO DATA

```python
#Filter out those rows which does not contain any data 
Womens_Shoes_df = Womens_Shoes_df.dropna() 

#Confirm rows are dropped
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### FILTER THE DATAFRAME AND DROP DUPLICATES, AND RE-ORDER

```python
# Mens DF - Select Highest Price ID and Drop Duplicates 

# convert prices column to Float 
Womens_Shoes_df['prices.amountMax'] = pd.to_numeric(Womens_Shoes_df['prices.amountMax'], errors = 'coerce')

# sort DF based on Price
Womens_Shoes_df = Womens_Shoes_df.sort_values('prices.amountMax', ascending=False)

# Remove Duplicate ID rows 
Womens_Shoes_df.drop_duplicates(subset=['id'], inplace = True, keep = 'first')

#Confirm rows are dropped
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### ADD A COLUMN TO THE DATAFRAME AND FILL WITH 'Mens'

```python
#Create an Gender column for Men's Shoes
Womens_Shoes_df['Gender'] = "Women's"

#Confirm rows are dropped
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
#Womens_Shoes_df
```

#### RENAME COLUMNS AND CORRECT THE INDEX FOR VISUAL PURPOSES

```python
#The original dataframe had an index, so re-arrange the index 
Womens_Shoes_df.index = [x for x in range(1, len(Womens_Shoes_df.values)+1)]
Womens_Shoes_df.index.name = 'Index'
Womens_Shoes_df
```

```python
#Rename the columns for visual purposes
Womens_Shoes_df = Womens_Shoes_df.rename(columns = {"id":"ID", "brand":"Brand", "name":"Name", "prices.amountMax":"Price($)", 
                                                    "prices.currency":"Currency", "prices.dateSeen":"Date"})

#Reorganise the columns for visualisation
Womens_Shoes_df = Womens_Shoes_df[['ID', 'Name', 'Brand', 'Colours', 'Gender', 'Price($)', 'Currency', 'Date']]

#Show the 'shape' of the dataframe
shape = Womens_Shoes_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```
